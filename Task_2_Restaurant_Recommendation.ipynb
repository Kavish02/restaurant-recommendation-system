# 🍽️ Restaurant Recommendation System

## Objective:
To build a restaurant recommendation system using content-based filtering based on user preferences like cuisine, price range, and location.

## Step 1: Data Preprocessing

The dataset is loaded and cleaned by removing missing values.
Important columns such as restaurant name, cuisines, city, price range, and ratings are selected.

```python
import pandas as pd

df = pd.read_csv("Dataset .csv")

df = df[['Restaurant Name', 'Cuisines', 'City', 'Price range', 'Aggregate rating']]
df = df.dropna()

df['Cuisines'] = df['Cuisines'].astype(str)
df['City'] = df['City'].astype(str)
df['Price range'] = df['Price range'].astype(str)

df.head()
```

## Step 2: Feature Engineering

Relevant features such as cuisines, city, and price range are combined into a single column.
This helps represent each restaurant as a single input for similarity comparison.

```python
df['features'] = df['Cuisines'] + " " + df['City'] + " " + df['Price range']
df[['Restaurant Name', 'features']].head()
```

## Step 3: Text Vectorization

TF-IDF (Term Frequency-Inverse Document Frequency) is used to convert text data into numerical form.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english')
feature_matrix = vectorizer.fit_transform(df['features'])
```

## Step 4: Similarity Calculation

Cosine similarity is used to measure similarity between restaurants.

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(feature_matrix)
```

## Step 5: Recommendation System

A function is created that takes user preferences and returns top 5 similar restaurants sorted by rating.

```python
def recommend_restaurants(user_input):
    user_vec = vectorizer.transform([user_input])
    scores = cosine_similarity(user_vec, feature_matrix)

    top_indices = scores[0].argsort()[-5:][::-1]

    result = df.iloc[top_indices][['Restaurant Name', 'Cuisines', 'City', 'Price range', 'Aggregate rating']]

    return result.sort_values(by='Aggregate rating', ascending=False)
```

## Step 6: Testing and Evaluation

The system is tested using different user preferences.

```python
print("🔹 Italian Delhi 2\n")
print(recommend_restaurants("Italian Delhi 2"))

print("\n🔹 Chinese Mumbai 1\n")
print(recommend_restaurants("Chinese Mumbai 1"))

print("\n🔹 North Indian Pune 3\n")
print(recommend_restaurants("North Indian Pune 3"))
```

## Conclusion

The restaurant recommendation system successfully provides personalized suggestions
based on cuisine, location, and price range using content-based filtering.
